```python
# CHOOSING LEARNING RATE


import json 
import matplotlib.pyplot as plt

learning_curve_001 = json.load(open('learning_curves/0.01.json'))
learning_curve_005 = json.load(open('learning_curves/0.05.json'))
learning_curve_02 = json.load(open('learning_curves/0.2.json'))
```

```python
# train_losses = learning_curve_001['train_losses']
# valid_losses = learning_curve_001['valid_losses']
# learning_rate = 0.01

plt.figure(figsize=(10, 5))
plt.plot(range(len(learning_curve_001['train_losses'])), learning_curve_001['train_losses'], 'r-')
plt.plot(range(len(learning_curve_001['valid_losses'])), learning_curve_001['valid_losses'], 'r:')
plt.plot(range(len(learning_curve_005['train_losses'])), learning_curve_005['train_losses'], 'g-*')
plt.plot(range(len(learning_curve_005['valid_losses'])), learning_curve_005['valid_losses'], 'g:*')
plt.plot(range(len(learning_curve_02['train_losses'])), learning_curve_02['train_losses'], 'b-')
plt.plot(range(len(learning_curve_02['valid_losses'])), learning_curve_02['valid_losses'], 'b:')
plt.xlabel("Number of epochs")
plt.ylabel("Loss")
plt.title(f"MobileNet: Train & Validation Loss vs Number of epochs")
plt.legend(['Train 0.01', 'Valid 0.01', 'Train 0.05', 'Valid 0.05', 'Train 0.2', 'Valid 0.2'])
plt.savefig(f'images/losses_learning_rates.pdf')
plt.show()

```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 12))

# Plot for losses
ax1.plot(range(len(learning_curve_001['train_losses'])), learning_curve_001['train_losses'], 'r-')
ax1.plot(range(len(learning_curve_001['valid_losses'])), learning_curve_001['valid_losses'], 'r:')
ax1.plot(range(len(learning_curve_005['train_losses'])), learning_curve_005['train_losses'], 'g-*')
ax1.plot(range(len(learning_curve_005['valid_losses'])), learning_curve_005['valid_losses'], 'g:*')
ax1.plot(range(len(learning_curve_02['train_losses'])), learning_curve_02['train_losses'], 'b-')
ax1.plot(range(len(learning_curve_02['valid_losses'])), learning_curve_02['valid_losses'], 'b:')
ax1.set_xlabel("Number of epochs")
ax1.set_ylabel("Loss")
ax1.set_title("MobileNet: Train & Validation Loss vs Number of epochs")
ax1.legend(['Train 0.01', 'Valid 0.01', 'Train 0.05', 'Valid 0.05', 'Train 0.2', 'Valid 0.2'])

# Plot for accuracies
ax2.plot(range(len(learning_curve_001['train_accuracies'])), learning_curve_001['train_accuracies'], 'r-')
ax2.plot(range(len(learning_curve_001['valid_accuracies'])), learning_curve_001['valid_accuracies'], 'r:')
ax2.plot(range(len(learning_curve_005['train_accuracies'])), learning_curve_005['train_accuracies'], 'g-*')
ax2.plot(range(len(learning_curve_005['valid_accuracies'])), learning_curve_005['valid_accuracies'], 'g:*')
ax2.plot(range(len(learning_curve_02['train_accuracies'])), learning_curve_02['train_accuracies'], 'b-')
ax2.plot(range(len(learning_curve_02['valid_accuracies'])), learning_curve_02['valid_accuracies'], 'b:')
ax2.set_xlabel("Number of epochs")
ax2.set_ylabel("Accuracy")
ax2.set_title("MobileNet: Train & Validation Accuracy vs Number of epochs")
ax2.legend(['Train 0.01', 'Valid 0.01', 'Train 0.05', 'Valid 0.05', 'Train 0.2', 'Valid 0.2'])

plt.tight_layout()
plt.savefig(f'images/losses_and_accuracies_learning_rates.pdf')
plt.show()
```
