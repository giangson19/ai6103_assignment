# CHOOSING LEARNING RATE

```python



import json 
import matplotlib.pyplot as plt

learning_curve_001 = json.load(open('learning_curves/0.01.json'))
learning_curve_005 = json.load(open('learning_curves/0.05.json'))
learning_curve_02 = json.load(open('learning_curves/0.2.json'))
```

```python
# train_losses = learning_curve_001['train_losses']
# valid_losses = learning_curve_001['valid_losses']
# learning_rate = 0.01

plt.figure(figsize=(10, 5))
plt.plot(range(len(learning_curve_001['train_losses'])), learning_curve_001['train_losses'], 'r-')
plt.plot(range(len(learning_curve_001['valid_losses'])), learning_curve_001['valid_losses'], 'r:')
plt.plot(range(len(learning_curve_005['train_losses'])), learning_curve_005['train_losses'], 'g-*')
plt.plot(range(len(learning_curve_005['valid_losses'])), learning_curve_005['valid_losses'], 'g:*')
plt.plot(range(len(learning_curve_02['train_losses'])), learning_curve_02['train_losses'], 'b-')
plt.plot(range(len(learning_curve_02['valid_losses'])), learning_curve_02['valid_losses'], 'b:')
plt.xlabel("Number of epochs")
plt.ylabel("Loss")
plt.title(f"MobileNet: Train & Validation Loss vs Number of epochs")
plt.legend(['Train 0.01', 'Valid 0.01', 'Train 0.05', 'Valid 0.05', 'Train 0.2', 'Valid 0.2'])
plt.savefig(f'images/losses_learning_rates.pdf')
plt.show()

```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 12))

# Plot for losses
ax1.plot(range(len(learning_curve_001['train_losses'])), learning_curve_001['train_losses'], 'r-')
ax1.plot(range(len(learning_curve_001['valid_losses'])), learning_curve_001['valid_losses'], 'r:')
ax1.plot(range(len(learning_curve_005['train_losses'])), learning_curve_005['train_losses'], 'g-*')
ax1.plot(range(len(learning_curve_005['valid_losses'])), learning_curve_005['valid_losses'], 'g:*')
ax1.plot(range(len(learning_curve_02['train_losses'])), learning_curve_02['train_losses'], 'b-')
ax1.plot(range(len(learning_curve_02['valid_losses'])), learning_curve_02['valid_losses'], 'b:')
ax1.set_xlabel("Number of epochs")
ax1.set_ylabel("Loss")
ax1.set_title("MobileNet: Train & Validation Loss vs Number of epochs")
ax1.legend(['Train 0.01', 'Valid 0.01', 'Train 0.05', 'Valid 0.05', 'Train 0.2', 'Valid 0.2'])

# Plot for accuracies
ax2.plot(range(len(learning_curve_001['train_accuracies'])), learning_curve_001['train_accuracies'], 'r-')
ax2.plot(range(len(learning_curve_001['valid_accuracies'])), learning_curve_001['valid_accuracies'], 'r:')
ax2.plot(range(len(learning_curve_005['train_accuracies'])), learning_curve_005['train_accuracies'], 'g-*')
ax2.plot(range(len(learning_curve_005['valid_accuracies'])), learning_curve_005['valid_accuracies'], 'g:*')
ax2.plot(range(len(learning_curve_02['train_accuracies'])), learning_curve_02['train_accuracies'], 'b-')
ax2.plot(range(len(learning_curve_02['valid_accuracies'])), learning_curve_02['valid_accuracies'], 'b:')
ax2.set_xlabel("Number of epochs")
ax2.set_ylabel("Accuracy")
ax2.set_title("MobileNet: Train & Validation Accuracy vs Number of epochs")
ax2.legend(['Train 0.01', 'Valid 0.01', 'Train 0.05', 'Valid 0.05', 'Train 0.2', 'Valid 0.2'])

plt.tight_layout()
plt.savefig(f'images/losses_and_accuracies_learning_rates.pdf')
plt.show()
```

# Weight Decay

```python
import json 
import matplotlib.pyplot as plt

cosine_annealing_weight_decay_00001 = json.load(open('learning_curves/cosine_annealing_weight_decay_0.0001.json'))
cosine_annealing_weight_decay_00005 = json.load(open('learning_curves/cosine_annealing_weight_decay_0.0005.json'))
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 12))

# Plot for losses with weight decay 1e-4 and 5e-4
ax1.plot(range(len(cosine_annealing_weight_decay_00001['train_losses'])), cosine_annealing_weight_decay_00001['train_losses'], 'r-')
ax1.plot(range(len(cosine_annealing_weight_decay_00001['valid_losses'])), cosine_annealing_weight_decay_00001['valid_losses'], 'r:')
ax1.plot(range(len(cosine_annealing_weight_decay_00005['train_losses'])), cosine_annealing_weight_decay_00005['train_losses'], 'g-')
ax1.plot(range(len(cosine_annealing_weight_decay_00005['valid_losses'])), cosine_annealing_weight_decay_00005['valid_losses'], 'g:')
ax1.set_xlabel("Number of epochs")
ax1.set_ylabel("Loss")
ax1.set_title("MobileNet: Train & Validation Loss vs Number of epochs (Weight Decay & Cosine Annealing)")
ax1.legend(['Train 1e-4', 'Valid 1e-4', 'Train 5e-4', 'Valid 5e-4'])

# Plot for accuracies with weight decay 1e-4 and 5e-4
ax2.plot(range(len(cosine_annealing_weight_decay_00001['train_accuracies'])), cosine_annealing_weight_decay_00001['train_accuracies'], 'r-')
ax2.plot(range(len(cosine_annealing_weight_decay_00001['valid_accuracies'])), cosine_annealing_weight_decay_00001['valid_accuracies'], 'r:')
ax2.plot(range(len(cosine_annealing_weight_decay_00005['train_accuracies'])), cosine_annealing_weight_decay_00005['train_accuracies'], 'g-')
ax2.plot(range(len(cosine_annealing_weight_decay_00005['valid_accuracies'])), cosine_annealing_weight_decay_00005['valid_accuracies'], 'g:')
ax2.set_xlabel("Number of epochs")
ax2.set_ylabel("Accuracy")
ax2.set_title("MobileNet: Train & Validation Accuracy vs Number of epochs (Weight Decay & Cosine Annealing)")
ax2.legend(['Train 1e-4', 'Valid 1e-4', 'Train 5e-4', 'Valid 5e-4'])

plt.tight_layout()
plt.savefig(f'images/losses_and_accuracies_weight_decay.pdf')
plt.show()
```

```python
# Extract final losses and accuracies for weight decay experiments
final_data_weight_decay = {
    'Weight Decay': ['1e-4', '5e-4'],
    'Final Train Loss': [
        cosine_annealing_weight_decay_00001['train_losses'][-1],
        cosine_annealing_weight_decay_00005['train_losses'][-1]
    ],
    'Final Valid Loss': [
        cosine_annealing_weight_decay_00001['valid_losses'][-1],
        cosine_annealing_weight_decay_00005['valid_losses'][-1]
    ],
    'Final Train Accuracy': [
        cosine_annealing_weight_decay_00001['train_accuracies'][-1],
        cosine_annealing_weight_decay_00005['train_accuracies'][-1]
    ],
    'Final Valid Accuracy': [
        cosine_annealing_weight_decay_00001['valid_accuracies'][-1],
        cosine_annealing_weight_decay_00005['valid_accuracies'][-1]
    ]
}

# Create DataFrame
final_df_weight_decay = pd.DataFrame(final_data_weight_decay)
final_df_weight_decay_transposed = final_df_weight_decay.set_index('Weight Decay').T

# Display the table in LaTeX format
print(final_df_weight_decay_transposed.to_latex(float_format="%.3f", caption="Final losses and accuracies for different weight decay values", label="tab:final_losses_accuracies_weight_decay", bold_rows=True))
```

```python
import json 
import matplotlib.pyplot as plt

# cosine_annealing_weight_decay_00001 = json.load(open('learning_curves/cosine_annealing_weight_decay_0.0001.json'))
cosine_annealing_weight_decay_00005_relu = json.load(open('learning_curves/cosine_annealing_weight_decay_0.0005.json'))
cosine_annealing_weight_decay_00005_sigmoid = json.load(open('learning_curves/cosine_annealing_weight_decay_0.0005_sigmoid.json'))
```
